```python
from time import process_time
from math import inf

def euclidian_graph(file : list, graph_start : int):
    i = graph_start
    coords = list()
    try:
        while(file[i].isnumeric()):
            coords.append((int(file[i + 1]), int(file[i + 2])))
            i += 3
    except:
         while(file[i].isnumeric()):
            coords.append((float(file[i + 1]), float(file[i + 2])))
            i += 3

    graph = list()
    for x in range(len(coords)):
        node_distance = list()
        for y in range(len(coords)):
            if x == y:
                node_distance.append(inf)
            else:
                node_distance.append(pow(pow(coords[x][0] - coords[y][0], 2) + pow(coords[x][1] - coords[y][1], 2), 0.5))
        graph.append(tuple(node_distance))

    return tuple(graph)


def file_reader(path : str):
    with open(path, 'r') as file:
        file = file.read()
        file = file.split()
    return euclidian_graph(file, file.index('NODE_COORD_SECTION') + 1)

def distance_verify(graph, traveled):
    distance = 0
    for i in range(len(traveled) - 1):
        distance += graph[traveled[i]][traveled[i + 1]]
    distance += graph[traveled[-1]][traveled[0]]
    return distance
    
```

# Guloso

```python
def greedy_TSP(graph):
    traveled = [0]
    distance = 0
    current = 0
    begin = process_time()

    while(len(traveled) != len(graph)):
        min_distance = inf
        min_index = inf

        for index in range(len(graph)):
            if not index in traveled and graph[current][index] < min_distance and round(graph[current][index], 2) > 0 and round(current, 2) != index:
                min_distance = graph[current][index]
                min_index = index
        current = min_index
        traveled.append(current)
        distance += min_distance
    distance += graph[current][0]
    
    return distance, process_time() - begin, traveled
```

# 2 OPT

```python
def two_opt(graph, traveled):
    best_solution = traveled
    best_distance = distance_verify(graph, traveled)
    for i in range(len(traveled)):
        for j in range(i + 1, len(traveled)):
            new_traveled = traveled.copy()
            new_traveled[i:j] = reversed(new_traveled[i:j])
            new_distance = distance_verify(graph, new_traveled)
            if new_distance < best_distance:
                best_distance = new_distance
                best_solution = new_traveled
    return best_solution, best_distance
```

```python
def verify_distance(graph : list, route : list) -> float:
    dist = 0
    current = 0

    for i in range(1, len(graph)):
        dist += graph[current][route[i]]
        current = route[i]

    dist += graph[current][0]
    
    return dist


def two_opt(graph : list, route : list) -> tuple:
    start = process_time()
    best_route = route
    improved = True
    while(improved):
        improved = False
        for i in range(1, len(route) - 1):
            for j in range(i + 1, len(route)):
                if j - i != 1:
                    new_route = route[:]
                    new_route[i:j] = route[j-1:i-1:-1]
                    if verify_distance(graph, new_route) < verify_distance(graph, best_route):
                        best_route = new_route
                        improved = True

        route = best_route
    
    return best_route, verify_distance(graph, best_route), process_time() - start
import numpy as np


```

```python
from random import choice

def random_greedy(graph : list, alfa : float) -> tuple:
    start_time = process_time()

    current = 0
    route = [current]
    distance = 0

    while(len(route) != len(graph)):
        line = graph[current]
        
        available_nodes = [(node, index) for node, index in zip(line, range(len(line))) if index not in route and node < inf]

        available_nodes.sort()

        # print(available_nodes)

        max_distance = 0

        # for i in range(len(available_nodes)):
        #     if line[available_nodes[i]] > max_distance:
        #         max_distance = line[available_nodes[i]]

        gama = int(alfa * len(available_nodes))
        # print(gama)

        restrict_values = list()

        for i in range(gama):
            restrict_values.append(available_nodes[i][1])

        # restrict_values = [node for node in available_nodes if line[node] <= gama]

        # print(restrict_values)
        if len(restrict_values) > 0:
            # print('pode ir para algum')
            travel_to = choice(restrict_values)

        else:
            # print('remediando')
            nodes = [i for _, i in available_nodes]
            travel_to = choice(nodes)

        # print(current)
        # print(travel_to)
        distance += graph[current][travel_to]
        route.append(travel_to)
        current = travel_to

    distance += graph[current][0]
    route.append(0)

    return distance, route, process_time() - start_time

def grasp(graph : list, alfa : float) -> tuple:
    '''
    Calculates a random greedy output and executes a two opt swap on it
    @param: graph
    @param: alfa
    @return: route
    @return: distance
    '''
    _, route, _ = random_greedy(graph, alfa)
    route, dist, _ = two_opt(graph, route)
    return route, dist
```
